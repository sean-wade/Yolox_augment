```python
import torch
from yolox.models import YOLOXHeadAttr, YOLOXHead, YOLOPAFPN, YOLOX
from yolox.data.datasets import DTLD_ATTRIBUTES

from yolox.utils import get_model_info
```

```python
dummy_input = torch.rand((1,3,640,640))
```

```python
backbone = YOLOPAFPN(1, 1, in_channels=[256, 512, 1024])
head = YOLOXHead(1, 1, in_channels=[256, 512, 1024])
model = YOLOX(backbone, head)

print(get_model_info(model, (640, 640)))

model.eval()
res = model(dummy_input)
print(res)
print(res.shape)
```

```python
backbone2 = YOLOPAFPN(1, 1, in_channels=[256, 512, 1024])
head2 = YOLOXHeadAttr(1, 1, in_channels=[256, 512, 1024], attribute_dict=DTLD_ATTRIBUTES)
model2 = YOLOX(backbone2, head2)

print(get_model_info(model2, (640, 640)))
model2.eval()
res2 = model2(dummy_input)
print(res2)
print(res2.shape)
```
